### Visualising the Target Variable (`SalePrice`)

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['SalePrice'], kde=True)
plt.title("Distribution of Sale Prices")
plt.xlabel("SalePrice")
```

The `SalePrice` is skewed to the right due to a relatively small number of high-value homes.

```python
import numpy as np

df['LogSalePrice'] = np.log1p(df['SalePrice'])
sns.histplot(df['LogSalePrice'], kde=True)
plt.title("Log-Transformed Sale Price")
```

The distribution is less skewed after applying a simple log tranformation.

### Check for Missing Values

```python
import pandas as pd

# Calculate missing values and percentages
missing_data = df.isnull().sum()
missing_percentage = (missing_data / len(df)) * 100

missing_info = pd.DataFrame({
    'Missing Values': missing_data, 
    'Percentage': missing_percentage
})

missing_info['dtype'] = df.dtypes[missing_info.index]

# Filter and sort by percentage missing
missing_info = missing_info[missing_info['Missing Values'] > 0]
missing_info = missing_info.sort_values(by='Percentage', ascending=False)

print(missing_info)

# Reorder columns in df to match sorted missing_info
df_nulls = df[missing_info.index]

# Keep only columns with more than 5 nulls
df_nulls_filtered = df_nulls.loc[:, df_nulls.isnull().sum() > 5]

# Plot heatmap
sns.heatmap(df_nulls_filtered.isnull().T, cbar=False)
```

```python
# Separate categorical (object) and numerical (float64) columns
categorical_cols = df_nulls.select_dtypes(include='object').columns
numerical_cols = df_nulls.select_dtypes(include='float64').columns

# Plot countplots of categorical columns
plt.figure(figsize=(16, len(categorical_cols) * 3))

for idx, col in enumerate(categorical_cols, 1):
    plt.subplot(len(categorical_cols), 4, idx)
    sns.countplot(data=df_nulls, x=col)
    plt.xticks(rotation=45)
    plt.title(col)

plt.tight_layout()
plt.show()

# Plot histograms of numerical columns
plt.figure(figsize=(16, len(numerical_cols) * 3))

for idx, col in enumerate(numerical_cols, 1):
    plt.subplot(len(numerical_cols), 4, idx)
    sns.histplot(data=df_nulls, x=col, kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```
